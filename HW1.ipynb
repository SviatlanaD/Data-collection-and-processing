```python
import requests
import json
from pprint import pprint
```

```python
params = {'q': 'name', 'id': '100750236'}

url = 'https://api.github.com'
user = 'SviatlanaD'

responce = requests.get(f'{url}/users/{user}/repos')

with open('data.json', 'w') as f:
    json.dump(responce.json(), f)

for i in responce.json():
    print(i['full_name'])
```
